```python
import numpy as np
import pandas as pd
import collections as collect
import os
```

```python
popSize=10000
```

```python
def genomeCounter(WT,popSize,*genomeCounts): 
    genomeCounts=dict()
    for H in WT.host.unique():
        genomeCounts[H]={}
        for S in WT.set.unique():
            genomeCounts[H][S]={} 
            for passage in WT.passage.unique():
                print(H+"_"+S+"_"+passage)
                inputDF=WT.loc[(WT.set==S)&(WT.host==H)&(WT.passage==passage),:]
                genomes=runSampling(H,S,passage,popSize,inputDF)
                print("Tabulating...")
                genomeCounts[H][S][passage]=[collect.Counter(genome) for genome in genomes]
                print("\t...done.")
    return(genomeCounts)
```

```python
def runSampling(H,S,passage,popSize,inputDF):
    colnames=inputDF.columns
    pArray=np.array(inputDF.iloc[:,4:],dtype=float)
    popSample=np.array(stateSampler(pArray,popSize,colnames))
    genomes=popSample.transpose()
    return(genomes)
```

```python
def stateSampler(inputMatrix,N,colnames):
    print("Sampling "+str(popSize)+" genomes...")
    choice=[np.random.choice(a=colnames[4:], p=Xi,size=N) for Xi in inputMatrix] #position
    print("\t...done.\n")
    return choice
```

```python
infile="/Users/ptdolan/RSandbox/Wtable.csv"
outfile="/Users/ptdolan/RSandbox/sampledGenomes.csv"

WT=pd.read_csv(infile)
# Run counter function on table
genomeCounts=genomeCounter(WT,popSize)#Run function to sample and count genomes
# Reformat output counts
dictPanda=[ for passage in genomeCounts[H][S].keys() for S in genomeCounts[H].keys() for H in genomeCounts.keys()]
#dictPanda=[pd.DataFrame.from_dict(genomeCounts[H][S][passage]) for passage in genomeCounts[H][S].keys() for S in genomeCounts[H].keys() for H in genomeCounts.keys()]
names=[H+"_"+S+"_"+passage for passage in genomeCounts[H][S].keys() for S in genomeCounts[H].keys() for H in genomeCounts.keys()]

for i in range(len(names)):
    dictPanda[i]["name"] = names[i]
PD=pd.concat(dictPanda,sort=True)

# Write out to csv
PD.to_csv(outfile)
```
